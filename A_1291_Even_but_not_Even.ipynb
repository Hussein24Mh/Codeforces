# A. Even But Not Even

https://codeforces.com/problemset/problem/1291/A

```python
final = []
cases = int(input())
for i in range(0,cases):
#-----------------------------------------------------------------------------    
    xx , number = input() , str(input())
    
    sumodds = 0
    
    for i in range(1,11,2):
        sumodds += number.count(str(i))
            
    if sumodds <= 1:
        final.append(-1)
        continue
#-----------------------------------------------------------------------------
# try to make the sum of digites even
    le = len(number) 
    if sumodds %2 !=0:
        for i in range(1,le+1):
            if int(number[le-i]) %2 !=0 :
                number = number[0:le-i]+ number[le-i+1:]
                sumodds -= 1
                break
#-----------------------------------------------------------------------------
# remove last digit if it was even number
    while int(number[-1]) %2 == 0 or number[-1] == '0' : 
        number = number[:-1]
#-----------------------------------------------------------------------------
# finaly chech our case
    if sumodds %2 == 0 and int(number[-1])%2 != 0:
        final.append(int(number))
    else:final.append(-1)
#-----------------------------------------------------------------------------
# Print solution
for i in range(0,len(final)):
    print(final[i])
```
