# A. Even But Not Even

https://codeforces.com/problemset/problem/1291/A

```python
final = []
cases = int(input())
for i in range(0,cases):
    num_len, passed_str = input(), input()
    removednums = []
    
    list_num = [int(x) for x in str(passed_str)]
    
    while list_num[0] == 0:
        list_num.pop(0)
        if 0 not in removednums :removednums.append(0)

    while list_num[-1] %2 == 0 and len(list_num)>2:
        if list_num[-1] not in removednums: removednums.append(list_num.pop())
        else:list_num.pop()
            
    le = len(list_num)
    
    if sum(list_num) %2 !=0 and len(list_num)>2:
        for i in range(2,le+1):
            if list_num[le-i] %2 !=0:
                list_num.pop(le-i)
                break
                
    check = all(item in list_num for item in removednums)
    
    if list_num[-1] %2 != 0 and sum(list_num) %2 ==0 and check == True:
        final.append(''.join(str(e) for e in list_num))
    else:final.append(-1)
    
for i in final:
    print(i)
```
