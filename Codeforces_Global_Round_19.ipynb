# C. Andrew and Stones => Pass

```python
x = int(input())
for i in range(0,x):
    n, s = int(input()),list(map(int, input().split()))
    if n > 2 :
        evens = 0
        oddss = 0
        steps = 0
        odds1 = 0
        for j in range(1,n-1):
            if s[j]%2 == 0 and s[j] != 0:
                evens +=1
                steps += s[j]/2
            if s[j]%2 !=0 and s[j]>1:
                oddss +=1
                steps += (s[j]+1)/2
            if s[j] == 1:
                odds1 +=1
                steps += (s[j]+1)/2
        if (evens == 0 and oddss == 0) or (evens == 0 and n<4):
            print(-1)
            continue
        print(int(steps))
    else: print(0)
```
